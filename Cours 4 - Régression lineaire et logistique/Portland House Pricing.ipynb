```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_portland = pd.read_csv('Portland.csv')
X = df_portland['Size']
y = df_portland['Price']
```

```python
df_portland.head()
```

```python
plt.scatter(X, y, s=75, edgecolor='k')
plt.xlabel('Size', fontsize=18)
plt.ylabel('Price', fontsize=18)
plt.xlim(0, 5000)
plt.title('Portland house pricing')
plt.tight_layout()
plt.savefig('Portland dataset.jpg', dpi=450)
```

```python
from sklearn.linear_model import LinearRegression
lrgs =LinearRegression().fit(X.to_numpy().reshape(-1, 1), y)
```

```python
X_test = np.arange(0, 5000).reshape(-1, 1)
y_test = lrgs.predict(X_test)
```

```python
plt.plot(X_test, y_test, color='r', linewidth=3)
plt.scatter(X, y, s=75, edgecolor='k')
plt.xlabel('Size', fontsize=18)
plt.ylabel('Price', fontsize=18)
plt.xlim(0, 5000)
plt.title('Portland house pricing')
plt.tight_layout()
plt.savefig('Portland linear regression.jpg', dpi=450)
```
